##Task-1

-->Generate 100 random numbers in range 0 to 10 using numpy
randint function with the seed set to 10.

```python
import numpy as np

# Generate random numbers using numpy with seed set to 10
np.random.seed(10)
random_numbers = np.random.randint(0, 11, 100, dtype=int)
print(random_numbers)
```

-->Create a RDD using the paral-
lelize function using data generated in previous step.

```python
from pyspark import SparkContext

#SparkContext
sc = SparkContext("local", "RDD Task 1")

# Create RDD using parallelize function
# parallelize function is used to create an RDD
rdd = sc.parallelize(random_numbers)

```

-->Calculate the fre-quency of each number (0 - 10) using appropriate function of RDD.

```python
# Calculate the frequency of each number (0 - 10)
freq_count = rdd.countByValue()

# Print the frequency of each number
for num, count in sorted(freq_count.items()):
    print(f"Number {num}: {count} occurrences")

# Stop the SparkContext
sc.stop()
```

##TASK-2:

```python
from pyspark import SparkContext

sc = SparkContext("local", "WordFrequencyApp")

#Purpose: Loads the text8 dataset into an RDD.
text_data = sc.textFile("/content/drive/MyDrive/DATASCIENCE/sem2/BD/pyspark/text8")


# Purpose: Splits each line into words and flattens the result.
words_rdd = text_data.flatMap(lambda line: line.split(" "))

# Purpose: Filters words that contain the letter 'a'.
filtered_words_rdd = words_rdd.filter(lambda word: 'a' in word)


# Purpose: Maps each word to a key-value pair, where the word is the key and 1 is the value.
word_count_rdd = filtered_words_rdd.map(lambda word: (word, 1))

# The reduceByKey transformation aggregates the counts of each word by adding up the values associated with each key (word).
word_frequencies = word_count_rdd.reduceByKey(lambda x, y: x + y)

# Print the results
for word, frequency in sorted(word_frequencies.collect()):
    print(f"Word '{word}': {frequency} occurrences")

# Stop the SparkContext (important to do at the end of your Spark application)
sc.stop()
```

#• DataFrame Task:

```python
from pyspark.sql import SparkSession
from pyspark.sql.types import StructType, StructField, FloatType, StringType

spark = SparkSession.builder.master("local").appName('irisfile').getOrCreate()

# Load the iris JSON data into a DataFrames
df = spark.read.option("mode", "PERMISSIVE").json("/content/drive/MyDrive/DATASCIENCE/sem2/BD/pyspark/iris.json")
''' "PERMISSIVE" mode is used when encountering corrupt records (records that do not follow the expected schema). In this
mode, Spark tries to parse and load the data, and if it encounters corrupt records, it records them in a special column called _corrupt_record
and continues loading the rest of the data.'''


# Print the DataFrame schema
df.printSchema()

# Print a few lines of the raw JSON data
df.show(truncate=False)

# Define an explicit schema
custom_schema = StructType([
    StructField("petalLength", FloatType(), True),
    StructField("petalWidth", FloatType(), True),
    StructField("sepalLength", FloatType(), True),
    StructField("sepalWidth", FloatType(), True),
    StructField("species", StringType(), True)
])

# Load the iris JSON data into a DataFrame with the explicit schema
df_new = spark.read.schema(custom_schema).json("/content/drive/MyDrive/DATASCIENCE/sem2/BD/pyspark/iris.json")

# Print the DataFrame schema
df_new.printSchema()
df_new = df_new.dropna()

# Show the DataFrame
df_new.show()
```

```python
from pyspark.sql.functions import col


# Calculate Pearson Correlation between petalLength and petalWidth
correlation = df_new.stat.corr("petalLength", "petalWidth", method="pearson")
print(f"Pearson Correlation between petalLength and petalWidth: {correlation}")

# Show the columns sepalLength, sepalWidth, and species for rows where petalLength is >= 1.4
filtered_df = df_new.filter(col("petalLength") >= 1.4).select("sepalLength", "sepalWidth", "species")
filtered_df.show()

spark.stop()
```

-->the high value of 0.963 suggests a strong positive correlation between "petalLength" and "petalWidth." As "petalLength" increases, "petalWidth" tends to increase as well.
